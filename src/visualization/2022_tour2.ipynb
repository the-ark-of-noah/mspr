```python
import pandas as pd

df = pd.read_csv("../../data/processed/2022_elections_tour_2_cleaned.csv")

df
```

```python
import matplotlib.pyplot as plt

def create_pie_chart(data, title):
    # Couleurs personnalisées
    colors = ['#1f77b4', '#ff7f0e']

    # Explosion
    explode = (0.1, 0)

    # Créer le diagramme à secteurs
    fig, ax = plt.subplots()
    data.sum().plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, labels=data.columns)
    plt.title(title, fontsize=14)
    plt.show()

# Filtrer les données pour le département de 'yonne'
df_yonne = df.loc[df['code_postal'] == 89]

# Créer une nouvelle dataframe avec les colonnes 'Voix 1' et 'Voix 2' pour le département de 'yonne'
df_voix_yonne = df_yonne[['Voix 1', 'Voix 2']]

# Renommer les colonnes pour leur attribuer des libellés
df_voix_yonne = df_voix_yonne.rename(columns={'Voix 1': 'La République En Marche', 'Voix 2': 'Rassemblement National'})

# Créer le premier diagramme à secteurs pour le département de 'Yonne'
create_pie_chart(df_voix_yonne, 'Répartition des voix dans le département de Yonne')

# Créer une nouvelle dataframe avec les colonnes 'Voix 1' et 'Voix 2' pour tous les départements
df_voix_total = df[['Voix 1', 'Voix 2']]

# Renommer les colonnes pour leur attribuer des libellés
df_voix_total = df_voix_total.rename(columns={'Voix 1': 'La République En Marche', 'Voix 2': 'Rassemblement National'})

# Créer le deuxième diagramme à secteurs pour tous les départements
create_pie_chart(df_voix_total, 'Répartition des voix dans tous les départements')
```

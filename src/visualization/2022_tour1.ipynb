```python
import pandas as pd

df = pd.read_csv("../../data/processed/2022_elections_tour_1_cleaned.csv")

df
```

```python
import matplotlib.pyplot as plt

def create_pie_chart(data, title):
    # Trier les données par ordre croissant
    data = data.reindex(sorted(data.columns), axis=1)

    # Couleurs personnalisées
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#ff00ff', '#00ff00']

    # Explosion
    explode = tuple(0.1 for _ in range(len(data.columns)))

    # Créer le diagramme à secteurs
    data.sum().plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, labels=data.columns)
    plt.title(title)

    # Retourner l'axe
    return plt.gca()

# Créer une nouvelle dataframe avec les colonnes 'Voix 1' à 'Voix 12' pour le département de 'Yonne'
df_voix_yonne = df[df['code_postal'] == 89][['Voix 1', 'Voix 2', 'Voix 3', 'Voix 4', 'Voix 5', 'Voix 6', 'Voix 7', 'Voix 8', 'Voix 9', 'Voix 10', 'Voix 11', 'Voix 12']]
df_voix_yonne = df_voix_yonne.rename(columns={'Voix 1': 'Lutte Ouvrière', 'Voix 2': 'Parti Communiste Français', 'Voix 3': 'La République En Marche', 'Voix 4': 'Résistons', 'Voix 5': 'Rassemblement National', 'Voix 6': 'Reconquête', 'Voix 7': 'La France Insoumise', 'Voix 8': 'Parti Socialiste', 'Voix 9': 'Europe Écologie Les Verts', 'Voix 10': 'Les Républicains', 'Voix 11': 'Nouveau Parti Anticapitaliste', 'Voix 12': 'Debout La France'})

# Créer une nouvelle figure pour le premier diagramme
plt.figure(figsize=(6, 6))
# Créer le premier diagramme à secteurs pour le département de 'Yonne'
ax1 = create_pie_chart(df_voix_yonne, 'Répartition des voix dans le département de Yonne')

# Créer une nouvelle dataframe avec les colonnes 'Voix 1' à 'Voix 12' pour tous les départements
df_voix_total = df[['Voix 1', 'Voix 2', 'Voix 3', 'Voix 4', 'Voix 5', 'Voix 6', 'Voix 7', 'Voix 8', 'Voix 9', 'Voix 10', 'Voix 11', 'Voix 12']]
df_voix_total = df_voix_total.rename(columns={'Voix 1': 'Lutte Ouvrière', 'Voix 2': 'Parti Communiste Français', 'Voix 3': 'La République En Marche', 'Voix 4': 'Résistons', 'Voix 5': 'Rassemblement National', 'Voix 6': 'Reconquête', 'Voix 7': 'La France Insoumise', 'Voix 8': 'Parti Socialiste', 'Voix 9': 'Europe Écologie Les Verts', 'Voix 10': 'Les Républicains', 'Voix 11': 'Nouveau Parti Anticapitaliste', 'Voix 12': 'Debout La France'})

# Créer une nouvelle figure pour le deuxième diagramme
plt.figure(figsize=(6, 6))
# Créer le deuxième diagramme à secteurs pour tous les départements
ax2 = create_pie_chart(df_voix_total, 'Répartition des voix dans tous les départements')

# Afficher les diagrammes
plt.show()
```
